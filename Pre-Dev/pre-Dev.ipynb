```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load Data
train_x = np.load('/home/spokencall/data/data4Tensor/train_x.npy')
train_y = np.load('/home/spokencall/data/data4Tensor/train_y.npy')
test_x = np.load('/home/spokencall/data/data4Tensor//test_x.npy')
test_y = np.load('/home/spokencall/data/data4Tensor/test_y.npy')
test_label = np.load('/home/spokencall/data/data4Tensor/originalY/test_y.npy')
```

```python
# fix random seed for reproducibility
seed = 7
np.random.seed(seed)
```

```python
removedFromTrain = 0
clean_train_x = np.array([]).reshape(0, 10)
clean_train_y = np.array([]).reshape(0, 3)

for idx, array in enumerate(train_x):
    shouldBeRemoved = False
    for data in array:
        if np.isinf(data):
            shouldBeRemoved = True
    if shouldBeRemoved:
        removedFromTrain += 1
    else:
        clean_train_x = np.vstack((clean_train_x, array))
        clean_train_y = np.vstack((clean_train_y, train_y[idx])) 

print('Finish cleaning Train Data.')
print('Removed datapoints:  ' + str(removedFromTrain))
print('length of train_x:   ' + str(len(clean_train_x)))
print('length of train_y:   ' + str(len(clean_train_y)))


clean_test_x = np.array([]).reshape(0, 10)
clean_test_y = np.array([]).reshape(0, 3)
clean_label = np.array([]).reshape(0, 1)
removedFromTest = 0
for idx, array in enumerate(test_x):
    shouldBeRemoved = False
    for data in array:
        if np.isinf(data):
            shouldBeRemoved = True
        
    if shouldBeRemoved:
        removedFromTest += 1
    else:
        clean_test_x = np.vstack((clean_test_x, array))
        clean_test_y = np.vstack((clean_test_y, test_y[idx]))
        clean_label = np.vstack((clean_label, test_label[idx]))

print('Finish cleaning Test Data.')
print('Removed datapoints:   ' + str(removedFromTest))
print('length of test_x:   ' + str(len(clean_test_x)))
print('length of test_y:   ' + str(len(clean_test_y)))
print('length of labels:   ' + str(len(clean_label)))
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scaled_train_x = sc.fit_transform(clean_train_x)
scaled_test_x = sc.transform(clean_test_x)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Initializing Neural Network
classifier = Sequential()

# Adding the input layer and the first hidden layer
classifier.add(Dense(8,activation = 'relu', input_dim = 10))

# Adding the output layer
classifier.add(Dense(3, activation = 'softmax'))
```

```python
# Compiling Neural Network
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting our model 
classifier.fit(scaled_train_x, clean_train_y, batch_size = 5, epochs = 200)
```

```python
# Creating prediction
y_pred = classifier.predict(scaled_test_x)
```

```python
output = np.array([]).reshape(0, 3)

for data in y_pred:
    prop1 = data[0]
    prop2 = data[1]
    prop3 = data[2]
    if prop1 > prop2 > prop3 or  prop1 > prop3 > prop2:
        output = np.vstack((output, [1, 0,0]))
    elif prop2 > prop1 > prop3 or prop2 > prop3 > prop1:
        output = np.vstack((output, [0, 1,0]))
    elif prop3 > prop1 > prop2 or prop3 > prop2 > prop1:
        output = np.vstack((output, [0, 0,1]))

output[78]
```

```python
# [1 0 0] -> lang & me
# [0 1 0] -> me
# [0 0 1] -> nothing
CR, CA, PFA, GFA, FR, k = 0, 0, 0, 0, 0, 3.0
for idx, data in enumerate(output):
    # the students answer is correct in meaning and language
    # the system says the same -> accept
    if clean_label[idx] == 1 and np.array_equal(data,[1, 0, 0]):
        CA += 1
    # the system says correct meaning wrong language -> reject
    elif clean_label[idx] == 1 and np.array_equal(data,[0, 1, 0]):
        FR += 1
    # the system says incorrect meaning and incorrect language -> reject
    elif clean_label[idx] == 1 and np.array_equal(data,[0, 0, 1]):
        FR += 1
        
    # students answer is correct in meaning and wrong in language
    #The system says the same -> reject
    elif clean_label[idx] == 2 and np.array_equal(data,[0, 1, 0]):
        CR += 1
    # the system says correct meaning and correct language -> accept
    elif clean_label[idx] == 2 and np.array_equal(data,[1, 0, 0]):
        PFA += 1
    # the system says incorrect meaning and incorrect language -> reject
    elif clean_label[idx] == 2 and np.array_equal(data,[0, 0, 1]):
        CR += 1
    
    # students answer is incorrect in meaning and incorrect in language
    # the system says the same -> reject
    elif clean_label[idx] == 3 and np.array_equal(data,[0, 0, 1]):
        CR += 1
    # the system says correct meaning correct language -> accept
    elif clean_label[idx] == 3 and np.array_equal(data,[1, 0, 0]): 
        GFA += 1
    # the system says correct meaning incorrect language -> reject
    elif clean_label[idx] == 3 and np.array_equal(data,[0, 1, 0]):
        CR += 1

FA = PFA + k * GFA
Correct = CA + FR
Incorrect = CR + GFA + PFA
IncorrectRejectionRate = CR / ( CR + FA + 0.0 )
CorrectRejectionRate = FR / ( FR + CA + 0.0 )
print('\nINCORRECT UTTERANCES (' + str(Incorrect) + ')' )
print('CorrectReject    ' + str(CR) )
print('GrossFalseAccept ' + str(GFA) + '*' + str(k) + ' = ' + str(GFA * k) )
print('PlainFalseAccept ' + str(PFA) )
print('RejectionRate    ' + "{:.3f}".format(IncorrectRejectionRate) )

print('\nCORRECT UTTERANCES (' + str(Correct) + ')')
print('CorrectAccept    ' + str(CA) )
print('FalseReject      ' + str(FR) )
print('RejectionRate    ' + "{:.3f}".format(CorrectRejectionRate) )
print('\nD              ' +  "{:.3f}".format(IncorrectRejectionRate / CorrectRejectionRate) )
```
